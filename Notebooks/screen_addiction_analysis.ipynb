Statistical Summary


```python
stats_summary=df.describe()
print(stats_summary)
```

```python
app_usage=df["App Usage Time (min/day)"].values
```

```python
avg_app_usage=round(np.mean(app_usage),2)
print("Average App Usage (min/day):", avg_app_usage)
```

```python
max_app_usage=np.max(app_usage)
print("Highest usage: ",max_app_usage,"min/day")
min_app_usage=np.min(app_usage)
print("Lowest usage: ",min_app_usage,"min/day")
```

```python
std_app_usage=round(np.std(app_usage))
print("Standard Deviation: ",std_app_usage)
```

```python
above_avg_count=np.sum(app_usage > avg_app_usage)
print("Users Above Average App Usage:", above_avg_count)
```

```python
avg_screen_time=df['Screen On Time (hours/day)'].mean()
print("Average screen on time :",avg_screen_time)
```

```python
gender_count=df['Gender'].value_counts()
print("Number of male and female users :\n",gender_count)
```

```python
os_wise_distribution=df['Operating System'].value_counts()
print("OS wise distribution:\n",os_wise_distribution)
```

```python
os_distribution=df["Operating System"].value_counts()
print("Operating System Distribution: ")
print(os_distribution)
```

```python
avg_usage_by_os=df.groupby("Operating System")["App Usage Time (min/day)"].mean()
print("Average App usage by operating system:")
print(avg_usage_by_os)
```

```python
behavior_class_count=df["User Behavior Class"].value_counts()
print("User Behavior Class Distribution:")
print(behavior_class_count)
```

```python
addicted_users=df[df["User Behavior Class"]==4]
print("Simple Addicted Users:")
print(addicted_users.head())
```

```python
avg_battery_by_class=df.groupby("User Behavior Class")["Battery Drain (mAh/day)"].mean()
print("Average Battery Drain By Behavior Class:")
print(avg_battery_by_class)
      
```

```python
high_screen_users=df[df["Screen On Time (hours/day)"]>6]
high_screen_count=high_screen_users.shape[0]
print("Number of Users with Screen Time > 6 hours:", high_screen_count)
```

```python
correlation_matrix=df.select_dtypes(include=np.number).corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
top_app_users=df.sort_values(
    by="App Usage Time (min/day)",
    ascending=False
).head()

print("Top 5 users by App Usage: ")
print(top_app_users)
```

```python
avg_screen_by_age=df.groupby("Age")["Screen On Time (hours/day)"].mean()
print("Age-wise Average Screen Time:")
print(avg_screen_by_age)
```

```python
df["Usage Level"]=np.where(
    df["App Usage Time (min/day)"]>300,
    "High Usage",
    "Normal Usage"
)
print("Usage Level Column Added")
    
```

```python
usage_level_count=df["Usage Level"].value_counts()
print("Usage Level Distribution:")
print(usage_level_count)
```

Average App Usage by Age Group

```python
age_groups=pd.cut(
    df["Age"],
    bins=[18,25,35,45,60],
    labels=["18-25","26-35","36-45","46-60"]
)
avg_usage_by_age_group=(
    df.groupby(age_groups,observed=False)
        ["App Usage Time (min/day)"].mean().reset_index(drop=True)
)
print("Average App Usage By Age Group:")
print(avg_usage_by_age_group)
```

Addiction Percentage in Dataset

```python
total_users=df.shape[0]
addicted_count=df[df["User Behavior Class"]==4].shape[0]
addiction_percentage=(addicted_count/total_users)*100
print("Addicted Percentage: ",round(addiction_percentage,2),"%")
```

High App usage vs High Data Usage

```python
high_app_high_data=df[
    (df["App Usage Time (min/day)"]>300)&
    (df["Data Usage (MB/day)"]>1000)
    ]
print("Users with High App Usage and High Data Usage:")
print(high_app_high_data.shape[0])
```
