```python
df_describe=df.describe()
print(df_describe)
```

```python
app_usage=df["App Usage Time (min/day)"].to_numpy()
```

```python
mean_app_usage=np.mean(app_usage)
print("Mean : ",mean_app_usage)
```

```python
max_app_usage=np.max(app_usage)
print("Highest usage: ",max_app_usage,"min/day")
min_app_usage=np.min(app_usage)
print("Lowest usage: ",min_app_usage,"min/day")
```

```python
standard_deviation_app_usage=np.std(app_usage)
print("Standard Deviation: ",standard_deviation_app_usage)
```

```python
above_avg_users=app_usage>np.mean(app_usage)
sum_above_avg_users=np.sum(above_avg_users)
print("Count of users who use mobile more than average: ",sum_above_avg_users)
```
